```python
from ultralytics import YOLO

# Load pretrained YOLOv8 model
model = YOLO("yolo11n.pt")  # small, fast model

# Train the model
model.train(
    data="dataset3/data.yaml",  # path to your dataset yaml
    epochs=20,                 # number of training epochs
    imgsz=640,                 # image size
    batch=16,                  # adjust if GPU memory is low
    device=0                   # 0 = GPU, "cpu" if no GPU
)

# After training, best model is saved automatically
print("Training complete.")
```
